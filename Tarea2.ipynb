```python

# Connect to the database
conn = sqlite3.connect('Tarea.db')
c = conn.cursor()  # Create a cursor object

# Drop table if it exists - rep
c.execute("DROP TABLE IF EXISTS rep")


c.execute('''
    CREATE TABLE rep (
        id_school INTEGER PRIMARY KEY ,
        school_name TEXT
    )
''')


# Drop table if it exists - sc
c.execute("DROP TABLE IF EXISTS sc")

c.execute('''
    CREATE TABLE sc (
        id_school INTEGER PRIMARY KEY ,
        school_name TEXT,
        country TEXT
    )
''')

# Create table - td
c.execute("DROP TABLE IF EXISTS td")
c.execute('''
        CREATE TABLE td(
        id_school INTEGER,
        world_rank text,
        university_name TEXT,
        country TEXT,
        teaching FLOAT,
        research FLOAT,
        citations FLOAT,
        total_score FLOAT,
        num_students FLOAT,
        student_staff_ratio float,
        international_students FLOAT,
        year integer,
        FOREIGN KEY (id_school) REFERENCES rep(id_school)
    )
''')


c.execute("DROP TABLE IF EXISTS Shanghai")
c.execute('''
    CREATE TABLE Shanghai (
        id_school INTEGER,
        university_name TEXT,
        world_rank TEXT,
        national_rank INTEGER,
        alumni FLOAT,
        award FLOAT,
        hici FLOAT,
        ns FLOAT,
        year integer,
        FOREIGN KEY (id_school) REFERENCES rep(id_school)
    )
''')

# Create table - cw
c.execute("DROP TABLE IF EXISTS cw")
c.execute('''
    CREATE TABLE cw (
        id_school INTEGER,
        institution TEXT,
        world_rank TEXT,
        country TEXT,
        national_rank INTEGER,
        quality_of_education INTEGER,
        alumni_employment INTEGER,
        publications INTEGER,
        influence INTEGER,
        citations INTEGER,
        patents INTEGER,
        score FLOAT,
        year integer,
        FOREIGN KEY (id_school) REFERENCES rep(id_school)
    )
''')
# Create table -
c.execute("DROP TABLE IF EXISTS rankings")
c.execute('''
    CREATE TABLE rankings (
        id_school INTEGER,
        school_name TEXT,
        td_country TEXT,
        td_world_rank integer,
        shanghai_world_rank integer,
        cwur_world_rank integer,
        td_num_students FLOAT,
        td_international_students FLOAT,
        s_award FLOAT,
        cw_alumni_employment INTEGER,
        cw_publications INTEGER,
        cw_influence INTEGER,
        year integer,
        FOREIGN KEY (id_school) REFERENCES rep(id_school)
    )
''')
# Commit the changes
conn.commit()
conn.close()


```

```python
import pandas as pd
import os
import sqlite3

# Conectar a la base de datos
conn = sqlite3.connect('Tarea.db')

# Definir el directorio que contiene los archivos CSV
csv_dir = '/content/drive/My Drive/Tarea/'

# Diccionario que mapea archivos CSV a sus respectivas tablas
csv_to_table_mapping = {
    'cwurData - Repertory.csv': 'rep',
    'cwurData - School and Country.csv': 'sc',
    'cwurData - Shangai Data.csv': 'Shanghai',
    'cwurData - Times Data.csv': 'td',
    'cwurData - cwurData.csv': 'cw'
}

# Iterar a través de los archivos CSV e insertar datos en las tablas correspondientes
for csv_file, table_name in csv_to_table_mapping.items():
    # Definir la ruta completa al archivo CSV
    csv_file_path = os.path.join(csv_dir, csv_file)

    # Verificar si el archivo CSV existe
    if os.path.exists(csv_file_path):
        # Leer el archivo CSV en un DataFrame
        data_frame = pd.read_csv(csv_file_path)

        # Insertar el DataFrame en la tabla correspondiente
        data_frame.to_sql(table_name, conn, if_exists='append', index=False)
    else:
        print(f"File {csv_file} does not exist in the directory.")

# Confirmar los cambios y cerrar la conexión
conn.commit()
conn.close()



```

```python


conn = sqlite3.connect('Tarea.db')
c = conn.cursor()

# Insert data into rankings
c.execute('''INSERT INTO  rankings(
        id_school,
        school_name ,
        td_country,
        td_world_rank,
        shanghai_world_rank ,
        cwur_world_rank ,
        td_num_students,
        td_international_students,
        s_award,
        cw_alumni_employment,
        cw_publications,
        cw_influence,
        year
)
SELECT DISTINCT
        rep.id_school,
        rep.school_name,
        td.country,
        CASE WHEN (LOWER(td.world_rank) = "none") THEN "Did not participate" ELSE td.world_rank END,
        CASE WHEN (LOWER(shanghai.world_rank) = "none") THEN "Did not participate" ELSE shanghai.world_rank END,
        CASE WHEN (LOWER(cw.world_rank) = "none") THEN "Did not participate" ELSE cw.world_rank END,
        td.num_students,
        td.international_students,
        shanghai.award,
        cw.alumni_employment,
        cw.publications,
        cw.influence,
        td.year
FROM
    rep
LEFT JOIN
    td ON td.id_school = rep.id_school
LEFT JOIN
    shanghai ON shanghai.id_school = rep.id_school
LEFT JOIN
    cw ON cw.id_school = rep.id_school
    ''')
# Commit the changes
conn.commit()
# Now, select the distinct records

c.execute('''
    SELECT  distinct*
    from rankings
     WHERE rankings.year = "2012"
''')

# Fetch the results
results = c.fetchall()

# Print the results
for row in results:
    print(row)



```

```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect('Tarea.db')
c = conn.cursor()

try:
    c.execute('''
    WITH top_ten AS (
        SELECT DISTINCT
            rk.id_school,
            rk.school_name,
            CASE WHEN LOWER(rk.td_world_rank) < '10' THEN 1 ELSE 0 END AS td_top_ten,
            CASE WHEN LOWER(rk.shanghai_world_rank) < '10' THEN 1 ELSE 0 END AS sh_top_ten,
            CASE WHEN LOWER(rk.cwur_world_rank) < '10' THEN 1 ELSE 0 END AS cwur_top_ten,
            rk.year
        FROM rankings rk
        WHERE rk.year >= 2012
    ),

    bq AS (
        SELECT DISTINCT
            id_school,
            school_name,
            SUM(td_top_ten) AS td_top_ten_times,
            SUM(sh_top_ten) AS sh_top_ten_times,
            SUM(cwur_top_ten) AS cwur_top_ten_times,
            year
        FROM top_ten
        GROUP BY id_school, school_name
    )

    SELECT DISTINCT
        id_school,
        school_name,
        td_top_ten_times,
        sh_top_ten_times,
        cwur_top_ten_times,
        CASE
            WHEN (td_top_ten_times > sh_top_ten_times AND td_top_ten_times > cwur_top_ten_times) THEN 'stands out in Times Data'
            WHEN (sh_top_ten_times > td_top_ten_times AND sh_top_ten_times > cwur_top_ten_times) THEN 'stands out in Shanghai'
            WHEN (cwur_top_ten_times > td_top_ten_times AND cwur_top_ten_times > sh_top_ten_times) THEN 'stands out in CWUR'
            ELSE 'Draw'
        END AS Best_qualifier
    FROM bq
    WHERE year >= 2012
    ORDER BY id_school ASC;
    ''')

    # Fetch the results
    res = c.fetchall()

    # Check if results were returned
    if res:
        # Print the results
        print("University Rankings:")
        print("ID, School Name, TD Top Ten Times, Shanghai Top Ten Times, CWUR Top Ten Times, Best Qualifier")
        for row in res:
            print(row)
    else:
        print("No results found.")

except sqlite3.Error as e:
    print(f"An error occurred: {e}")

```
